### Q3. Write a Python function to calculate the probability density function of a normal distribution with given mean and standard deviation at a given point.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stat
import pylab 

data = np.linspace(0, 50, 50)
data
```

```python
def myplot(data):
    plt.figure(figsize=(12, 5))
    plt.subplot(1,2,1)
    sns.histplot(data, kde=True)
    plt.subplot(1,2,2)
    stat.probplot(data, dist="norm", plot=pylab)
    plt.show()
    
myplot(data)
```

```python
mean = data.mean()
std = data.std()
(mean, std)
```

```python
probability_pdf = stat.norm.pdf(50, loc=mean, scale=std)
probability_pdf
```

### Q4. What are the properties of Binomial distribution? Give two examples of events where binomial distribution can be applied.

Properties of Binomial Distribution:
1. Experiments includes "n" idendtical trials
2. There are only Two outcomes - Success S and Failure F.
3. Probability of Sucess is "p" throughout the experiment.
4. Probability of Failure is "q" throughout the experiment which is q = 1-p;


Examples:
1. Rolling a dice 10 times and the probability of getting 1 is 1/6th.
2. Tossing a coin.

### Q5. Generate a random sample of size 1000 from a binomial distribution with probability of success 0.4 and plot a histogram of the results using matplotlib.

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

X = np.random.binomial(n = 100, p =0.4, size=1000)

plt.hist(X)
plt.xlabel("Numbers")
plt.ylabel("Occurance")
```

### Q6. Write a Python function to calculate the cumulative distribution function of a Poisson distribution with given mean at a given point.

```python
from scipy.stats import poisson

k = np.arange(0, 15)

data = poisson.cdf(k ,mu=3)
data = np.round(data, 3)
plt.plot(k, data)
plt.show()
```

### Q7. How Binomial distribution different from Poisson distribution?

In a Binomial distribution, there is a fixed number of trials where as in Poisson there are infinite trials in given time interval.

### Q8. Generate a random sample of size 1000 from a Poisson distribution with mean 5 and calculate the sample mean and variance.

```python
from scipy.stats import poisson

X = np.random.poisson(lam = 0.5, size=1000)
print(f"Mean: {np.mean(X)}")
print(f"Variance: {np.var(X)}")
```
