### Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.

Lookup error is base class for the Index, Key errors and CodecRegistryError. Hierarchy is below

##### -- LookupError
##### ------- IndexError
##### ------- KeyError
##### ------------ NoSuchKernel
##### ------------ UnknownBackend
##### ------- CodecRegistryError


```python
## Index error
lst = [1,2,3,4,5,6]

try:
    for i in range(len(lst)+1):
        print(lst[i])
except IndexError as e:
    print("Lookup -- Index error:: ", e)
```

```python
## Key Error

my_dict = { "name": "Neil",
            "age": "31",
            "course_name": "DS Masters"}

try:
    user_input = input("What do you want?")
    print(f'My name is {my_dict["name"]}, and my {user_input} is {my_dict[user_input]}')
except KeyError as e:
    print("This is my KeyError: ", e)
```
